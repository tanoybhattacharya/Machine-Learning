```python
import numpy as np
import pandas as pd 

# Load the dataset
data = pd.read_csv("OnlineNewsPopularity.csv")
data.head()
```

```python
# The statistics of original target 
popularity_raw = data[data.keys()[-1]]
popularity_raw.describe()
```

__Encode the label by threshold 1400 Beacause majority 50% stock price over 1400 as per statistic__

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
popular_label = pd.Series(label_encoder.fit_transform(popularity_raw>=1400))
```

```python
# Features from dataset
features_raw = data.drop(['url',data.keys()[1],data.keys()[-1]], axis=1)
features_raw.head()
```

```python
#Get the statistic for all The Feature
features_raw.describe()
```

__Need data normalization for all the numeric feature__ 

```python
# Normalize the numerical features
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical = ['n_tokens_title', 'n_tokens_content', 'num_hrefs', 'num_self_hrefs', 'num_imgs','num_videos','average_token_length','num_keywords','self_reference_min_shares','self_reference_max_shares','self_reference_avg_sharess']
features_raw[numerical] = scaler.fit_transform(data[numerical])
features_raw.head()
```

```python
features_raw.describe()
```

__Now all feature attribute is now normalized with evidence of std__

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(features_raw,popular_label,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,VotingClassifier,AdaBoostClassifier
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
```

```python
lg = LogisticRegression()
rf = RandomForestClassifier()
ad = AdaBoostClassifier(n_estimators=50,learning_rate=1)
xg = XGBClassifier()
```

```python
# Accuray of Logistic regrassion
accuracy_score(lg.fit(X_train,y_train).predict(X_test),y_test)
```

```python
# Accuray of RandomSeacrh
accuracy_score(rf.fit(X_train,y_train).predict(X_test),y_test)
```

```python
# Accuray of Adaboost
accuracy_score(ad.fit(X_train,y_train).predict(X_test),y_test)
```

```python
# Accuray of XGBoost
accuracy_score(xg.fit(X_train,y_train).predict(X_test),y_test)
```

```python
# Accuray of VootingClassifier
vt = VotingClassifier(estimators = [('lg', lg),('rf', rf),('ad', ad),('xg',xg)],
                                       voting = 'hard')
accuracy_score(vt.fit(X_train,y_train).predict(X_test),y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(vt, features_raw,popular_label, cv=5)
scores
```
